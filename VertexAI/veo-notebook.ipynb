```python
# Paste the following into a new code block & run it to set the Google Cloud project information and create a client.
import os

PROJECT_ID = "qwiklabs-gcp-03-efa39c5c688b"
LOCATION = os.environ.get("GOOGLE_CLOUD_REGION", "us-central1")

client = genai.Client(vertexai=True, project=PROJECT_ID, location=LOCATION)
```

```python
# Next, define the helper functions to display media.
def show_video(gcs_uri):
    file_name = gcs_uri.split("/")[-1]
    !gsutil cp {gcs_uri} {file_name}
    media.show_video(media.read_video(file_name), height=500)


def display_images(image) -> None:
    fig, axis = plt.subplots(1, 1, figsize=(12, 6))
    axis.imshow(image)
    axis.set_title("Starting Image")
    axis.axis("off")
    plt.show()
```

```python
# Load the video generation model by running the following command.
video_model = "veo-2.0-generate-001"
```

## Task 2. Generate videos from a text prompt

```python
# Run the following code snippet to generate a video from a text prompt.
prompt = "a back cat reading a book"
aspect_ratio = "16:9"
output_gcs = "gs://qwiklabs-gcp-03-efa39c5c688b"

operation = client.models.generate_videos(
    model=video_model,
    prompt=prompt,
    config=types.GenerateVideosConfig(
        aspect_ratio=aspect_ratio,
        output_gcs_uri=output_gcs,
        number_of_videos=1,
        duration_seconds=5,
        person_generation="dont_allow",
        enhance_prompt=True,
    ),
)

while not operation.done:
    time.sleep(15)
    operation = client.operations.get(operation)
    print(operation)

if operation.response:
    show_video(operation.result.generated_videos[0].video.uri)
```
